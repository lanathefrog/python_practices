5. (7б) Визначте функцію, яка перевіряє, чи є число простим, використовуючи документацію з завдання 4.

```python
def is_prime(n):
    """
    Checks if the given number is a prime number.

    Examples:
        >>> is_prime(8)
        False
        >>> is_prime(13)
        True
        >>> is_prime("trippi-troppi")
        TypeError

    Args:
        n (int): The number to check.

    Returns:
        result (bool): True if n is a prime number, False otherwise.

    Raises:
        TypeError: if the input is not an integer.
    """

    if not isinstance(n, int):
        raise TypeError("Input must be an integer")
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

try:
    print(is_prime(17))
    print(is_prime(20))
    print(is_prime("bombardilo-crocodilo"))
except TypeError as e:
    print(e)

print(is_prime.__doc__)
```

```python
def longest_common_prefix(str1, str2):
    """
        Returns the longest common prefix of two strings, case-insensitive.

        Examples:
            >>> longest_common_prefix('Love', 'lov')
            'lov'
            >>> longest_common_prefix('Penny', 'pennywise')
            'penny'

        Args:
            str1 (str): The first string to compare.
            str2 (str): The second string to compare.

        Returns:
            result (str): The longest common prefix of str1 and str2.

        Raises:
            TypeError: if one of the arguments is not a string.
        """
```

7. (9б) Реалізуйте функцію згідно документації із завдання 6, використовуючи її.

```python
def longest_common_prefix(str1, str2):
    """
        Returns the longest common prefix of two strings, case-insensitive.

        Examples:
            >>> longest_common_prefix('Love', 'lov')
            'lov'
            >>> longest_common_prefix('Penny', 'pennywise')
            'penny'

        Args:
            str1 (str): The first string to compare.
            str2 (str): The second string to compare.

        Returns:
            result (str): The longest common prefix of str1 and str2.

        Raises:
            TypeError: if one of the arguments is not a string.
        """

    if not isinstance(str1, str) or not isinstance(str2, str):
        raise TypeError("Both arguments must be strings.")

    str1 = str1.lower()
    str2 = str2.lower()

    result = ""
    for i in range(min(len(str1), len(str2))):
        if str1[i] == str2[i]:
            result += str1[i]
        else:
            break

    return result
try:
    print(longest_common_prefix('Cat', 'Caterpillar'))
    print(longest_common_prefix('Dog', 'Dogs'))
    print(longest_common_prefix('Window', "Winner"))
    print(longest_common_prefix('Cat', 123))
except TypeError as e:
    print(e)


print(longest_common_prefix.__doc__)
```

9. (12б) Використовуючи документацію із завдання 8, реалізуйте відповідну функцію.

```python
def calculate_monthly_payment(loan_amount, years, down_payment, is_privileged):
    """
    Calculates the monthly payment for the 'yeOselya' mortgage program based on user category.

    The privileged category receives a 3% annual interest rate.
    The general category receives a 7% annual interest rate.

    Privileged categories include:
        - Defense forces and their families
        - Medical workers of state or municipal health institutions
        - Pedagogical workers of state or municipal educational institutions
        - Scientific and scientific-pedagogical workers of state or municipal institutions

    General conditions for everyone:
        - Maximum loan term: 20 years
        - Minimum down payment: 20% of the total property cost

    Examples:
        >>> calculate_monthly_payment(1000000, 20, 200000, True)
        4431.92
        >>> calculate_monthly_payment(1000000, 20, 200000, False)
        7756.76

    Args:
        loan_amount (float): Total property cost in currency units.
        years (int): Loan term in years (maximum 20).
        down_payment (float): Amount of initial payment (minimum 20% of loan_amount).
        is_privileged (bool): True if user belongs to privileged category, False otherwise.

    Returns:
        result (float): Monthly payment amount rounded to 2 decimal places.

    Raises:
        ValueError: If down payment is less than 20% of loan_amount or loan term exceeds 20 years.
    """
    if down_payment < 0.2 * loan_amount:
        raise ValueError("Down payment must be at least 20% of the loan amount.")
    if years > 20:
        raise ValueError("Maximum loan term is 20 years.")

    if is_privileged:
      annual_rate = 0.03
    else:
      annual_rate = 0.07
    monthly_rate = annual_rate / 12
    credit_amount = loan_amount - down_payment
    total_months = years * 12
    rate_multiplier = (1 + monthly_rate) ** total_months
    monthly_payment = credit_amount * (monthly_rate * rate_multiplier) / (rate_multiplier - 1)
    return round(monthly_payment, 6)

print(calculate_monthly_payment(500000, 15, 100000, False))
print(calculate_monthly_payment(100000, 5, 20000, True))
try:
        print(calculate_monthly_payment(1000000, 20, 100000, False))  # will raise ValueError for down payment
except ValueError as e:
        print(e)

print(calculate_monthly_payment.__doc__)
```

10. (3б) Реалізуйте функцію, яка отримує значення зі словника за ключем та повертає його, а за умови відсутності ключа відловлює KeyError.

```python
def get_currency(countries, country):
    try:
        return countries[country]
    except KeyError:
        raise KeyError(f"Country '{country}' not found in the list.")

countries = {
    "USA": "USD",
    "Germany": "EUR",
    "Japan": "JPY",
    "Canada": "CAD",
    "India": "INR"
}

try:
    print(f"Currency of USA: {get_currency(countries, 'USA')}")
    print(f"Currency of Brazil: {get_currency(countries, 'Brazil')}")
except KeyError as e:
    print(e)
```

11. (3б) Реалізуйте функцію із завдання 10 без відловлювання помилки, однак за допомогою конструкції if-else.

```python
def get_currency(countries, country):
    if country in countries:
        return countries[country]
    else:
        return f"Country '{country}' not found in the list."

countries = {
    "USA": "USD",
    "Germany": "EUR",
    "Japan": "JPY",
    "Canada": "CAD",
    "India": "INR"
}

try:
    print(f"Currency of USA: {get_currency(countries, 'USA')}")
    print(f"Currency of Brazil: {get_currency(countries, 'Brazil')}")
except KeyError as e:
    print(e)
```

12. (3б) Реалізуйте функцію з завдання 10, використовуючи метод словника, який дозволяє зробити операцію навіть без if-else.

```python
def get_currency(countries, country):
    return countries.get(country, f"Country '{country}' not found in the list.")

countries = {
    "USA": "USD",
    "Germany": "EUR",
    "Japan": "JPY",
    "Canada": "CAD",
    "India": "INR"
}

try:
    print(f"Currency of USA: {get_currency(countries, 'USA')}")
    print(f"Currency of Brazil: {get_currency(countries, 'Brazil')}")
except KeyError as e:
    print(e)
```

13. (5б) Визначте функцію, яка обробляє значення, введене користувачем з терміналу, приводячи його до цілого числа. Використайте відловлення відповідної помилки. Наприклад, людина має ввести свій вік.

```python
def get_user_age_try_except():
    user_input = input("Please enter your age: ")
    try:
        age = int(float(user_input))
        print(f"Your age is {age}.")
    except ValueError:
        print("Invalid input. Please enter a valid age.")

get_user_age_try_except()
```

14. (5б) Визначте функцію з завдання 13, де використайте конструкцію if-else замість відловдення помилки.

15. (5б) Визначте функцію з завдання 13, враховуючи додаткове обмеження, що число має бути у діапазоні від 18 до 120 (повнолітня людина). Відловіть відповідні типи помилок.

```python
def get_user_age():
    user_input = input("Please enter your age: ")
    try:
        age = int(float(user_input))
        if 18 <= age <= 120:
            print(f"Your age is {age}.")
        else:
            raise ValueError("Age must be between 18 and 120.")
    except ValueError as ve:
        print(f"Invalid input: {ve}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

get_user_age()
```

16. (12б) Напишіть код, який би генерував максимальну кількість помилок різного типу.

```python
class Clothing:
    def __init__(self, type):
        self.type = type

clothing_item = Clothing("jacket")
try:
    print(clothing_item.color)  # AttributeError
except AttributeError as e:
    print(f"AttributeError: {e}")

clothes = ["hat", "scarf", "gloves"]
try:
    print(clothes[5])  # IndexError
except IndexError as e:
    print(f"IndexError: {e}")

wardrobe = {"shirt": 3, "pants": 2}
try:
    print(wardrobe["socks"])  # KeyError
except KeyError as e:
    print(f"KeyError: {e}")

try:
    price = int("twenty")  # ValueError
except ValueError as e:
    print(f"ValueError: {e}")

try:
    print(shirt)  # NameError
except NameError as e:
    print(f"NameError: {e}")

size = "M"
try:
    print("The size is: " + 42)  # TypeError
except TypeError as e:
    print(f"TypeError: {e}")

try:
    discount = 50 / 0  # ZeroDivisionError
except ZeroDivisionError as e:
    print(f"ZeroDivisionError: {e}")
```
